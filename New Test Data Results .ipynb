```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
```

```python
dataset3 = pd.read_excel(r"C:\Users\88001381\Desktop\test_t.xlsx")
```

```python
X_test = dataset2.iloc[:, :-1]  
y_test = dataset3.iloc[:, -1] 
```

```python
rf_classifier = RandomForestClassifier(n_estimators=5, criterion = 'entropy', random_state=50, max_depth = 22)
```

```python
rf_classifier.fit(X_test, y_test)
```

```python
y_pred = rf_classifier.predict(X_test)
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)
print(conf_matrix)
```

```python
from sklearn.metrics import precision_score, \
    recall_score, confusion_matrix, classification_report, \
    accuracy_score, f1_score
```

```python
print('Accuracy:', accuracy_score(y_test, y_pred))
print('F1 score:', f1_score(y_test, y_pred))
print('Recall:', recall_score(y_test, y_pred))
print('Precision:', precision_score(y_test, y_pred))
```
