```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
```

```python
dataset = pd.read_csv(r"C:\Users\88001381\Desktop\PV_Model_Data(1).csv")
```

```python
dataset1 = pd.read_csv(r"C:\Users\88001381\Desktop\PV_Model_Data(3).csv")
```

```python
X_train = dataset.iloc[:, :-1]  
y_train = dataset1.iloc[:, -1] 
```

```python
rf_classifier = RandomForestClassifier(n_estimators=100, criterion = 'entropy', random_state=0, max_depth = 100)
```

```python
rf_classifier.fit(X_train, y_train)
```

```python
y_pred = rf_classifier.predict(X_train)
```

```python
conf_matrix = confusion_matrix(y_train, y_pred)
print(conf_matrix)
```

```python
from sklearn.metrics import precision_score, \
    recall_score, confusion_matrix, classification_report, \
    accuracy_score, f1_score
```

```python
print('Accuracy:', accuracy_score(y_train, y_pred))
print('F1 score:', f1_score(y_train, y_pred))
print('Recall:', recall_score(y_train, y_pred))
print('Precision:', precision_score(y_train, y_pred))
```
